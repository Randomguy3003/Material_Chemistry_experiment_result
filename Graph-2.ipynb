Vẽ cho nhiều nhóm

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

#read data
data = ["Data/D01_Nhom_1.txt",
        "Data/D01_Nhom_2.txt", 
        "Data/D01_Nhom_3.txt", 
        "Data/D01_Nhom_4.txt", 
        "Data/D01_Nhom_5.txt",
        
        "Data/D01_Nhom_7.txt",
        "Data/D01_Nhom_8.txt",
        "Data/D01_Nhom_9.txt",
        "Data/D01_Nhom_10.txt",
        "Data/D01_Nhom_11.txt",
        "Data/D01_Nhom_12.txt"
       ]

#set colors
np.random.seed(42)
num_files = len(data)
colors = np.random.rand(num_files, 3)  # RGB values [0,1]

#graph drawing
plt.figure(figsize=(20, 15))

for file_path in data:
    # đọc từng dataset, gắn giá trị (mất góc tiếng anh r nên để v đại :))) )
    dataset = np.loadtxt(file_path, skiprows = 3)
    nm = dataset[:, 0]
    counts = dataset[:, 1]

    color = np.random.rand(3,)
    plt.plot(nm, 
             counts, 
             color = color, 
             linewidth = 2, 
             label = file_path.split('/')[-1].replace('.txt', '').replace('D01_', '')
            )

# labels and formatting
plt.title('')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (CPS)')
plt.legend(title="Files", fontsize=8)
plt.grid(True)
plt.tight_layout()
plt.show()
```
